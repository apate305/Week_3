6.What is the ___ of `horsepower`?
- sum
- min
- max
- median
- Optional: mode (May need to google help)

```python
#sum of horsepower

hp_sum = sum(HP)
print ('Sum of the array is ',hp_sum)

a = len(HP)

#min of horsepower
def Min(HP, a):
    A = HP[0]
    for i in range(1,a):
        A = min(A, HP[i])
    return A
print ("Minimum:", Min(HP, a))

# Max of horsepower
def Max(HP, a):
    B = HP[0]
    for i in range(1,a):
        B = max(B, HP[i])
    return B
print ("Max:", Max(HP, a))

#Median of horsepower
C = np.median(HP)
print("Median:", C)
```
